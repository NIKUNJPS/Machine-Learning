```python
#detection
Q3=df['Purchase'].quantile(0.75)
Q1=df['Purchase'].quantile(0.25)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
df[(df['Purchase']<LL)|(df['Purchase']>UL)]
```

```python
#removal of outliers

Q3=df['Purchase'].quantile(0.75)
Q1=df['Purchase'].quantile(0.25)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
#not outlier : x>=LL and x<=UL
df=df[(df['Purchase']>=LL)&(df['Purchase']<=UL)]
```
