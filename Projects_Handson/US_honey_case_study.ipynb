1)Which states are rarely contributing to honey production for the last 27 years ?

2) Which are the top 5 Honey producing states in US


```python
# prompt: 1)Which states are rarely contributing to honey production for the last 27 years ?
# 2) Which are the top 5 Honey producing states in US

# 1) Which states are rarely contributing to honey production for the last 27 years?

# Group by state and calculate the total production for each state
state_production = df.groupby('state')['production'].sum()

# Find states with low total production
low_production_states = state_production[state_production < state_production.quantile(0.25)].index.tolist()

print("States rarely contributing to honey production:")
print(low_production_states)


# 2) Which are the top 5 Honey producing states in US?

# Group by state and calculate the total production for each state
state_production = df.groupby('state')['production'].sum()

# Sort the states by total production in descending order and get the top 5
top_5_states = state_production.sort_values(ascending=False).head(5)

print("\nTop 5 Honey producing states:")
top_5_states
```
